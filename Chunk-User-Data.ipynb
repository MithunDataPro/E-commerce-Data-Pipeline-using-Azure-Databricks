```python
import os

# Define chunk size
chunk_size = len(df) // 5 + (1 if len(df) % 5 else 0)

# Define output directory (make sure it exists)
output_dir = r"M:\Projects\Databricks\Azure Databricks\jfreex-e-commerce-users-of-a-french-c2c-fashion-store\chunk-data"
os.makedirs(output_dir, exist_ok=True)

# Split the dataframe into chunks & save them (chunk6.csv to chunk10.csv)
for i in range(5):
    start = i * chunk_size
    end = min(start + chunk_size, len(df))  # Prevents index overflow
    chunk_df = df.iloc[start:end]
    
    # Save each chunk with index starting from 6 to 10
    chunk_file = os.path.join(output_dir, f"chunk{i+6}.csv")
    chunk_df.to_csv(chunk_file, index=False)

    print(f"Saved {chunk_file}")
```
